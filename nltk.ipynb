```python
corpus = open("corpus_teste.txt", "r").read()
```

## Exercício 1</br>
Escolha um corpus qualquer e, usando as funções do NLTK, faça:</br>
a. Tokenize o corpus inteiro 
(palavras, números e pontuações)</br>
b. Verifique a quantidade de tokens do corpus</br>
c. Tokenize o corpus apenas por suas palavras</br>
d. Verifique a quantidade de palavras do corpus</br>
e. Verifique a frequência de palavras no corpus</br>
f. Verifique quais são as 5, 10 e 15 palavras mais frequentes do corpus</br>
g. Extraia as stopwords do NLTK (não do corpus ainda)</br>
h. Verifique a frequência dos tokens sem stopwords do corpus</br>

```python
#a) Tokenize o corpus inteiro (palavras, números e pontuações)
all_tokens = nltk.word_tokenize(corpus)
```

```python
#b) Verifique a quantidade de tokens do corpus
print(f'Quantidade de tokens do corpus: {len(all_tokens)}')
```

```python
#c) Tokenize o corpus apenas por suas palavras
tokenizer = RegexpTokenizer(r'[A-z]\w*') #return word characters
word_tokens =  tokenizer.tokenize(corpus)
```

```python
#d) Verifique a quantidade de palavras do corpus
print(f'Quantidade de palavras do corpus: {len(word_tokens)}')
```

```python
#e) Verifique a frequência de palavras no corpus
frequencia = nltk.FreqDist(word_tokens)
frequencia.most_common()
```

```python
#f) Verifique quais são as 5, 10 e 15 palavras mais frequentes do corpus
frequencia.most_common(5)
frequencia.most_common(10)
frequencia.most_common(15)
```

```python
#g) Extraia as stopwords do NLTK (não do corpus ainda)
stopwords = nltk.corpus.stopwords.words('portuguese')
```

```python
#h) Verifique a frequência dos tokens sem stopwords do corpus
no_stopwords = [x for x in word_tokens if x.lower() not in stopwords]
freq_no_stopwords = nltk.FreqDist(no_stopwords)
freq_no_stopwords.most_common()
```
