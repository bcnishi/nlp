## Exercício 1
Utilizando um corpus de seu interesse (pode ser o que foi utilizado na aula) e utilizando o spaCy, faça:
- a) Tokenize todo o corpus;
- b) Tokenize todo o corpus e retorne a lista com strings;
- c) Tokenize todo o corpus e retorne apenas as palavras;
- d) Retorne a quantidade de palavras do corpus;
- e) Tokenize todo o corpus e retorne apenas os números;
- f) Retorne a quantidade de números do corpus;
- g) Tokenize todo o corpus e retorne apenas as pontuações;
- h) Retorne a quantidade de pontuações do corpus;
- i) Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!);
- j) Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus.

```python
corpus = open("corpus_teste.txt", "r", encoding="utf-8").read()
```

```python
import spacy

nlp = spacy.load("pt_core_news_lg")
doc = nlp(corpus)
```

```python
#a) Tokenize todo o corpus
tokens = [token for token in doc]
print(tokens)
```

```python
#b) Tokenize todo o corpus e retorne a lista com strings
tokens = [token.orth_ for token in doc]
tokens
```

```python
#c) Tokenize todo o corpus e retorne apenas as palavras
alpha_tk = [token.orth_ for token in doc if token.is_alpha]
print(f'Alpha tokens:\n')
alpha_tk
```

```python
#d) Retorne a quantidade de palavras do corpus
print(f'Quantidade de palavras do corpus: {len(alpha_tk)}')
```

```python
#e) Tokenize todo o corpus e retorne apenas os números
digit_tk = [token.orth_ for token in doc if token.is_digit]
```

```python
#f) Retorne a quantidade de números do corpus
print(f'Quantidade de números do corpus: {len(digit_tk)}')
```

```python
#g) Tokenize todo o corpus e retorne apenas as pontuações
punct_tk = [token.orth_ for token in doc if token.is_punct]
```

```python
#h) Retorne a quantidade de pontuações do corpus
print(f'Quantidade de pontuações do corpus: {len(punct_tk)}')
```

```python
#i) Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!)
from nltk import FreqDist
freq_punct = FreqDist(punct_tk)
freq_punct
```

```python
#j) Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus
whitespace = [token.orth_ for token in doc if token.is_space]
print(f'Quantidade de espaços do corpus: {len(whitespace)}')
```
