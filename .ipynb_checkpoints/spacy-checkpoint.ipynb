## Exercício 1
Utilizando um corpus de seu interesse (pode ser o que foi utilizado na aula) e utilizando o spaCy, faça:
- a) Tokenize todo o corpus;
- b) Tokenize todo o corpus e retorne a lista com strings;
- c) Tokenize todo o corpus e retorne apenas as palavras;
- d) Retorne a quantidade de palavras do corpus;
- e) Tokenize todo o corpus e retorne apenas os números;
- f) Retorne a quantidade de números do corpus;
- g) Tokenize todo o corpus e retorne apenas as pontuações;
- h) Retorne a quantidade de pontuações do corpus;
- i) Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!);
- j) Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus.

```python
corpus = open("corpus_teste.txt", "r", encoding="utf-8").read()
```

```python
import spacy

nlp = spacy.load("pt_core_news_lg")
doc = nlp(corpus)
```

```python
#a) Tokenize todo o corpus
tokens = [token for token in doc]
print(tokens)
```

```python
#b) Tokenize todo o corpus e retorne a lista com strings
tokens = [token.orth_ for token in doc]
tokens
```

```python
#c) Tokenize todo o corpus e retorne apenas as palavras
alpha_tk = [token.orth_ for token in doc if token.is_alpha]
print(f'Alpha tokens:\n')
alpha_tk
```

```python
#d) Retorne a quantidade de palavras do corpus
print(f'Quantidade de palavras do corpus: {len(alpha_tk)}')
```

```python
#e) Tokenize todo o corpus e retorne apenas os números
digit_tk = [token.orth_ for token in doc if token.is_digit]
```

```python
#f) Retorne a quantidade de números do corpus
print(f'Quantidade de números do corpus: {len(digit_tk)}')
```

```python
#g) Tokenize todo o corpus e retorne apenas as pontuações
punct_tk = [token.orth_ for token in doc if token.is_punct]
```

```python
#h) Retorne a quantidade de pontuações do corpus
print(f'Quantidade de pontuações do corpus: {len(punct_tk)}')
```

```python
#i) Retorne a frequência de pontuação do corpus (obs: pode usar o NLTK nesse exercício!)
from nltk import FreqDist
freq_punct = FreqDist(punct_tk)
freq_punct
```

```python
#j) Tokenize todo o corpus e retorne a quantidade de espaços presentes no corpus
whitespace = [token.orth_ for token in doc if token.is_space]
print(f'Quantidade de espaços do corpus: {len(whitespace)}')
```

## Execício 2
Ainda utilizando o mesmo corpus e o spaCy, faça:
- a) Analise o 1º parágrafo do corpus, lematizando todos os tokens possíveis. Verifique se foi retornado os lemas corretos;
- b) Ainda com o 1º parágrafo, retorne todas as classes gramaticais dos tokens;
- c) Retorne todas as classes gramaticais do corpus;
- d) Use o 1º parágrafo para retornar as informações morfológicas dos tokens;
- e) Retorne todas as entidades nomeadas do corpus;
- f) Verifique a classificação de cada entidade nomeada do corpus;
- g) Use o displaCy para mostrar graficamente todas as entidades nomeadas do corpus.

```python
#a) Analise o 1º parágrafo do corpus, lematizando todos os tokens possíveis. 
#Verifique se foi retornado os lemas corretos;

lemmas = [token.lemma_ for token in doc[8:24]]
lemmas
```

```python
#b) Ainda com o 1º parágrafo, retorne todas as classes gramaticais dos tokens

pos = [(token.orth_, token.pos_) for token in doc[8:24]]
pos
```

```python
#c) Retorne todas as classes gramaticais do corpus
pos2 = [(token.orth_, token.pos_) for token in doc]
pos2
```

```python
#d) Use o 1º parágrafo para retornar as informações morfológicas dos tokens
morfologicas = [(token.orth_, token.morph) for token in doc[8:24]]
morfologicas
```

```python
#e) Retorne todas as entidades nomeadas do corpus
entidades = list(doc.ents)
entidades
```

```python
#f) Verifique a classificação de cada entidade nomeada do corpus
detalhes_entidades = [(entidade, entidade.label_) for entidade in doc.ents]
detalhes_entidades
```

```python
#g) Use o displaCy para mostrar graficamente todas as entidades nomeadas do corpus
html = spacy.displacy.render(doc, style='ent')
output_path = open('entidades_nomeadas.html', 'w', encoding='utf-8')
output_path.write(html)
output_path.close()
```
